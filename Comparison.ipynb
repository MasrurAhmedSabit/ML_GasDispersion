```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from tensorflow.keras.layers.experimental.preprocessing import Normalization
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from data_preprocess import *
```

```python
df=load_roundhill_data()
test_set=[14,26,32,15,33]
X_train=df.drop(test_set)
X_test=df.loc[test_set]
Y_train=X_train.pop('C(mg/m3)')
Y_test=X_test.pop('C(mg/m3)')
scaler=MinMaxScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
r2,mae,mse=[],[],[]
```

```python
tr,ts,k=[],[],[]
for i in range(1,30):
    est=KNeighborsRegressor(n_neighbors=i).fit(X_train_scaled,Y_train)
    tr.append(est.score(X_train_scaled,Y_train))
    ts.append(est.score(X_test_scaled,Y_test))
    k.append(i)
df=pd.DataFrame({'n_neighbors':k,'Train Score':tr,'Test_score':ts}).set_index('n_neighbors')
b=df[df['Test_score']==df['Test_score'].max()].index[0]
knn=KNeighborsRegressor(n_neighbors=b).fit(X_train_scaled,Y_train)
r2.append(r2_score(Y_test.values,knn.predict(X_test_scaled)))
mae.append(mean_absolute_error(Y_test.values,knn.predict(X_test_scaled)))
mse.append(mean_squared_error(Y_test.values,knn.predict(X_test_scaled)))
```

```python
r2_s,mae_s,mse_s=[],[],[]
for i in range(10):
    rf=RandomForestRegressor(n_estimators=50,criterion='mae',max_features='auto',n_jobs=-1).fit(X_train_scaled,Y_train)
    r2_s.append(r2_score(Y_test.values,rf.predict(X_test_scaled)))
    mae_s.append(mean_absolute_error(Y_test.values,rf.predict(X_test_scaled)))
    mse_s.append(mean_squared_error(Y_test.values,rf.predict(X_test_scaled)))
r2.append(np.mean(r2_s))
mae.append(np.mean(mae_s))
mse.append(np.mean(mse_s))
```

```python
r2_s,mae_s,mse_s=[],[],[]
for i in range(10):
    normalizer=Normalization()
    normalizer.adapt(np.array(X_train))
    model=Sequential([
        normalizer,
        Dense(100,activation='relu'),
        Dense(100,activation='relu'),
        Dense(10,activation='relu'),
        Dense(1),
    ])
    model.compile(optimizer='adam',loss='mean_absolute_error')
    history=model.fit(X_train,Y_train,epochs=100,verbose=1,batch_size=500)
    r2_s.append(r2_score(Y_test.values,model.predict(X_test)))
    mae_s.append(mean_absolute_error(Y_test.values,model.predict(X_test)))
    mse_s.append(mean_squared_error(Y_test.values,model.predict(X_test)))
r2.append(np.mean(r2_s))
mae.append(np.mean(mae_s))
mse.append(np.mean(mse_s))
```

```python
RH_result=pd.DataFrame({'Model':['K_Neighbors','Random Forest','Neural Network'],'R$^2$':r2,'Mean Absolute Error':mae,'Mean Squared Error':mse}).set_index('Model')
RH_result
```
